```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
xvals = 2*x
```

```python
yinterp = np.interp(xvals, np.arange(len(x)), y)
```

```python
xvals
```

```python
plt.plot(yinterp, '-x')
```
